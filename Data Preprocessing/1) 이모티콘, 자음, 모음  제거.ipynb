```python
df_seoul['지역'] = '서울'
```

```python
df_jeju['지역'] = '제주'
```

```python
dfs = [df_seoul, df_jeju]
```

```python
df = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
df.head()
```

```python
df['평점'] = df['평점 및 날짜'].str.extract(r'(\d+)점')
df['날짜'] = df['평점 및 날짜'].str.extract(r'·\s*(.+)')
```

```python
df
```

```python
df = df.drop(['평점 및 날짜'], axis=1)
```

```python
df['가입기간'] = df['가입기간'].str.replace('에어비앤비 가입 기간 ', '', regex=False)
```

```python
df['번역여부'] = df['번역여부'].str.extract(r'^(.*?)에서 번역됨')
```

```python
df['호스트'] = df['호스트'].str.extract(r'^(.*?)님의 답글')
```

```python
df
```

```python
desired_order = ['id', '가입기간', '평점', '날짜', '리뷰', '번역여부', '호스트', '답글 날짜', '답글', '게스트 선호', '슈퍼 호스트', '카테고리', '지역']
df = df[desired_order]
```

```python
import re
def clean_text(text):
    if isinstance(text, str):
        text_no_emojis = emoji.replace_emoji(text, replace='')
        
        text_no_special_chars = re.sub(r'[^\w\s]', '', text_no_emojis)
        text_cleaned = re.sub(r'[ㅋㅎㅠ]+', '', text_no_special_chars) 

        text_hangul_only = re.sub(r'[^가-힣\s]', '', text_cleaned) 
        
        return text_hangul_only
    return text

df['리뷰'] = df['리뷰'].fillna('').apply(clean_text)
```
