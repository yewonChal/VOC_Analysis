```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

data = pd.read_csv('DATA/Jeju_morphs.csv')

korean_reviews = data[data['번역여부'].str.contains('한국어', na=False)] 
foreign_reviews = data[~data['번역여부'].str.contains('한국어', na=False)] 

korean_reviews['형태소'] = korean_reviews['형태소'].fillna('')
foreign_reviews['형태소'] = foreign_reviews['형태소'].fillna('')

korean_text = " ".join(str(review) for review in korean_reviews['형태소']) 

korean_wordcloud = WordCloud(width=800, height=400, background_color='white', font_path='/Library/Fonts/AppleGothic.ttf').generate(korean_text)

foreign_text = " ".join(str(review) for review in foreign_reviews['형태소']) 
foreign_wordcloud = WordCloud(width=800, height=400, background_color='white', font_path='/Library/Fonts/AppleGothic.ttf').generate(foreign_text)

plt.figure(figsize=(10, 6))
plt.imshow(korean_wordcloud, interpolation='bilinear')
plt.title('한국어 리뷰 워드 클라우드')
plt.axis('off')
plt.show()

plt.figure(figsize=(10, 6))
plt.imshow(foreign_wordcloud, interpolation='bilinear')
plt.title('외국어 리뷰 워드 클라우드')
plt.axis('off')
plt.show()
```

```python
data = pd.read_csv('DATA/Jeju_morphs.csv')
data.head(1)
```

```python
from collections import Counter
from wordcloud import WordCloud

def count_words(text):
    words = text.split()
    return Counter(words)

category_word_counts = {}
for category, group in data.groupby('카테고리'):
    word_counter = Counter()
    for review in group['형태소']:
        if isinstance(review, str):
            word_counter.update(count_words(review))
    category_word_counts[category] = word_counter
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

for category, word_counter in category_word_counts.items():
    most_common_words = word_counter.most_common(10) 
    words, counts = zip(*most_common_words)
    
    plt.figure(figsize=(10, 6))
    
    sns.barplot(x=list(counts), y=list(words), color="#FF5A5F")
    
    plt.title(f'{category} 카테고리에서 많이 나온 단어')
    plt.xlabel('빈도수')
    plt.ylabel('단어')
    plt.show()
```
